## 1b: Swarm Plot 

**INSTRUCTION: Demonstrate the age in both genders separeted by manner of death by a swarm plot.**

```python
# with seaborn
plt.figure(figsize=(9,6))
sns.swarmplot(x="gender",y="age",data=kill,hue="manner_of_death")
plt.show()
```

```python
o=kill.city.value_counts().head(12).index
o
```

```python
e=kill.city.value_counts().head(12)
e
```

```python
# with seaborn barplot
fig, ax=plt.subplots(figsize=(12,6))     #City kaybolmadi :)

sns.barplot(x=o,y=e,ax=ax)

ax.set_xticklabels(o, rotation=45)
ax.set_title("Most dangerous cities",color="b",fontweight="bold",size=20)
ax.set_ylabel("")

plt.show()

```

///

```python
# with seaborn countplot
fig, ax=plt.subplots(figsize=(12,6))     
sns.countplot(x="city",data=kill,order=o)
ax.set_xticklabels(o, rotation=45)
ax.set_title("Most dangerous cities",color="b",fontweight="bold",size=20)

plt.show()
```
