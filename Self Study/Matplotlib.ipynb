```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4,5,6,7]
y = [30,31,23,44,54,33,23]
```

```python
plt.plot(x,y);
```

```python
plt.plot(x,y,color="green");
```

```python
plt.plot(x,y,color="purple");
```

```python
plt.plot(x,y,color="purple",linewidth=5);
```

```python
plt.plot(x,y,color="yellow",linewidth=25);
```

```python
plt.plot(x,y,color="red",linewidth=3,lineStyle="dashdot");
```

```python
plt.plot(x,y,color="red",linewidth=3,lineStyle="dotted")
```

```python
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(x,y,color="red",linewidth=3,lineStyle="dotted")
```
