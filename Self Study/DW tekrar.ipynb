```python
####
np.linspace(0,10,10)
y1 = x
y2 = x**2
y3 = x**3
y4 = np.sqrt(x)

plt.figure()
plt.subplot(2,2,1)
plt.plot(x,y1,"ro")
plt.title("y1=x")

plt.subplot(2,2,2)
plt.plot(x,y2,"bo")
plt.title("$y2=x**2$")   

plt.subplot(2,2,3)
plt.plot(x,y3,"yo")
plt.title("$y2=x**3$")

plt.subplot(2,2,4)
plt.plot(x,y4,"go")
plt.title("$square root of x$")

plt.tight_layout()

plt.show()
```

```python
# possible linestype options --, , -., :, steps
fig, ax = plt.subplots(figsize=(12,6))
ax.plot(x, x-1, color="green", lw=3, linestyle='-') # solid
ax.plot(x, x-2, color="green", lw=3, ls='-.') # dash and dot
ax.plot(x, x-3, color="green", lw=5, ls=':') # dots
ax.plot(x, x-4, color="green", lw=3, ls='--') # dashes
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,6))
# Use marker for string code
# Use markersize or ms for size
ax.plot(x, x-1,marker='+',markersize=10)
ax.plot(x, x-2,marker='o',ms=15) #ms can be used for markersize
ax.plot(x, x-3,marker='s',ms=10,lw=0.5) # make linewidth zero to see only markers
ax.plot(x, x-4,marker='1',ms=10)
plt.show()
```

```python
# Plot Range with set_ylim & set_xlim
x = np.arange(1,10)
fig, ax = plt.subplots(nrows=1,ncols=2, figsize=(10,4))
ax[0].plot(x,x**2,x,x**3,x,x**4, 'r')


ax[1].plot(x,x**2,x,x**3,x,x**4, 'r')
ax[1].set_xlim([1,3])
ax[1].set_ylim([1,50])       #kivrimi ilkinde goremezken set_ylim ile yakinlastiriyoruz tabi caizse
plt.show()
```

```python
x = np.arange(1,10)
fig, ax = plt.subplots(nrows=1,ncols=2,figsize=(10,4))

ax[0].plot(x,x**2,x,x**3,x,x**4, "g")

ax[1].plot(x,x**2,x,x**3,x,x**4, "g")

ax[1].set_xlim([1,5])
ax[1].set_ylim([1,50])

ax[0].axvline(x=5,ls="--")   #X de axvline
ax[0].axhline(y=3000,ls="--")

ax[1].axhline(y=30,ls="--")   #Y da axhline

plt.show()
```
